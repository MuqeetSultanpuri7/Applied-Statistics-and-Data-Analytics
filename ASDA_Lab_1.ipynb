```python
import string
from collections import defaultdict
import pandas as pd

allowed_characters = string.ascii_letters + "'"

common_nouns = ["idea", "reaction", "term", "France", "Amar", "mid", "Limit", "degrowth", "capitalism", 
                "agenda", "model", "process", "methodology", "ecroissance", "GDP", "economics", "Georgesscu-Roegen"]

prepositions = ['in', 'on', 'at', 'by', 'with', 'about', 'against', 'among', 'between', 'through', 'during', 'before',
                'after', 'above', 'below', 'to', 'from', 'up', 'down', 'under', 'over', 'again', 'further', 'then', 'once']

# To check if a word is a noun
def is_noun(word):  
    if word in common_nouns:
        return True
    if word[0].isupper() and word.lower() not in ["a","an","the"] and word.lower() not in prepositions:
        return True
    if word.endswith(("tion", "ment", "ism", "ity")):
        return True
    return False

# To check if a word is an article
def is_article(word):
    if word.lower() in ["a","an","the"]:
        return True
    return False

# To check if a word is a preposition
def is_preposition(word):
    if word.lower() in prepositions:
        return True
    return False


```

```python
text = """ The idea of d'ecroidssance emerged in France in the mid-1970s as a dirrect reaction to The Limit to 
Growt(1972).According to Mumca(2013).d'ecroidssance was first used as a term by Amar(1973) in a rather loose way critically  
refelect upon the western model of econmic development, before  French translations of some of the works of Georgesscu-Roegen
(Georgesscu-Roegen, 1979) popularized d'ecroidssance as an alternative to the already estabilished idea of 
z egisrne (i.e zero growth).

In particular, Georgesscu-Roegen(1975) application of ther-modynamics to economic processes has been central to
d'ecroissance is its English speaking sibling of degrowth. A central characteristic of d'ecroidssance its critical stance 
towards the western model human development  (Martinez-Aller et al.,2010). Concretely, it questions its taclt assumption 
that economic development and human progress are correlated  (e.g., lllich, 1973, 1974; Latouche, 1986) and proposes 
voluntary simplicity as a means of free the individual from model consumerlsm, with places like Can Decreix at the 
French-Spanish border as exemplified alternative modes of living (see http://www.candecreix. cat/). Historically, the French 
speaking disoourse about d'ecroidssance is older than the one about degrowth, but French word d'ecroidssance is often 
translated as “degrowth” in English texts (Martinez-Aller et al.,2010), making it a de-facto synonym without actually 
meaning the exact same thing.

In comparison to d'ecroidssance, the degrowth discourse is less radical and rooted in ecological economics. Ecological 
economics criticize economic mainstream, for example the use of GDP as a policy indicator, but it seeks to indicate 
economics as a discipline rather than aban-doning it, in particular by adding to the body of ecological economic theory 
from which concrete policy proposals can be developed. Thus, degrowth aims at reforms of existing institutions and system 
rather than revolutions (Kallis, 2011)

Generally, we follow the research design and methodology estab-lished in our earlier systematic reviews that aimed at 
particular concepts (Abson et al.. 2014 Brandt et al.,2013; Nuchter et al., 2021) or research fields (Drupp et al., 2020). 
We outline this process in the following, a graphical overview can be found in Fig. 1.

According to Martinez-Aller et al.,2010, d'ecroissance has a clear eco-socialist agenda, i.e.  d'ecroidssance Sees economic 
questions as necessarily political questions, which ultimately lead to the abolishment of capitalism, Furthermore,  
d'ecroidssance is “anti-economics” in a sense that economics is seen as a discipline that should be abandoned (ibid.).
"""
```

```python
# To convert the text into the required format, access and assign the insights of each word
def process_text(text):

    word_details = defaultdict(lambda:{
        'frequency' : 0,
        'length': 0,
        'lowercase' : "",
        'is_noun' : False,
        'is_article' : False,
        'is_preposition' : False
    })


    cleaned_text = ""
    for char in text:
        if char in allowed_characters or char.isspace():
            cleaned_text+=(char)
        else:
            cleaned_text+=(" ")
    
    words = cleaned_text.split()
    
    # Assigning details of each word to the default dictionary
    for word in words: 
        lowercase_word = word.lower()
        is_noun_flag = is_noun(word)
        is_article_flag = is_article(word)
        is_preposition_flag = is_preposition(word)
        word_details[word]['frequency'] += 1
        word_details[word]['length'] = len(word)
        word_details[word]['lowercase'] = lowercase_word
        word_details[word]['is_noun'] = is_noun_flag
        word_details[word]['is_article'] = is_article_flag
        word_details[word]['is_preposition'] = is_preposition_flag


    return word_details
```

```python
# result
analysis = process_text(text)
```

```python

# Data in not readable format
data = {
    'word' : [],
    'frequency' : [],
    'length' : [],
    'lowercase' : [],
    'is nown' : [],
    'is article' : [],
    'is preposition' : []
}

for word, info in analysis.items():
    data["word"].append(word)
    data["frequency"].append(info["frequency"])
    data["length"].append(info["length"])
    data["lowercase"].append(info["lowercase"])
    data["is nown"].append(info["is_noun"])
    data["is article"].append(info["is_article"])
    data["is preposition"].append(info["is_preposition"])

```

```python
# Converting the data in a tabular format
df = pd.DataFrame(data)
df
```
